```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
np.random.rand(200)
x = np.zeroes(200)
x[1] = 1
x[2] = 2
x[3] = 2
x[4] = 4


target = 2.53
mult = target/x.sum()

x *= mult
```

```python
plt.title('Gibbs collected vs Timestep')
plt.ylabel('Gibbs Normalized to RDKit')
plt.xlabel('Timestep')

plt.plot(np.arange(200), np.cumsum(x))
```
